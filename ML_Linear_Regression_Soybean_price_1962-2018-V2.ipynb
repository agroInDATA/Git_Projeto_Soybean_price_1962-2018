# Loading Data

```python
df = pd.read_csv('Datasets/Soybean_Prices_1962-2018.csv')
```

```python
df.describe()
```

```python
df.head(n=20)
```

# Data preparation

```python
#features
x = df.drop(["Date",'bean_high'], axis = 1)

#response variable
y = df['bean_high'].copy()
```

```python
x_train, x_test, y_train, y_test = ms.train_test_split(x, y, test_size = 0.2, random_state = 42)
```

```python
#model description
model_lr = lm.LinearRegression()

#model training
model_lr.fit(x_train, y_train)
```

```python
#prediction trainning
pred_train = model_lr.predict(x_train)

#prediction test 
pred_test = model_lr.predict(x_test)
```

# Model Trainning

```python
pred_train[0:100]
```

# Performance metrics

```python
#trainning
mae_train = m.mean_absolute_error(y_train, pred_train)
mape_train = np.mean(np.abs((y_train - pred_train) / y_train ))

#test
mae_test = m.mean_absolute_error(y_test, pred_test )
mape_test = np.mean(np.abs((y_test - pred_test) / y_test ))
```
