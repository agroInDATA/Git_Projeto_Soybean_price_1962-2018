# Loading Data

```python
df = pd.read_csv('Datasets/Soybean_Prices_1962-2018.csv')
```

```python
df.describe()
```

```python
df.head(n=20)
```

# Data preparation

```python
#features
x_train = df.drop(["Date",'bean_high'], axis = 1)

#response variable
y_train = df['bean_high'].copy()
```

```python
#model description
model_lr = lm.LinearRegression()

#model training
model_lr.fit(x_train, y_train)
```

```python
pred = model_lr.predict(x_train)
```

# Model Trainning

```python
pred[0:100]
```

# Performance metrics

```python
df_copy = df.copy()
```

```python
df_copy['prediction'] = pred
```

```python
df_copy['error'] = df_copy['bean_high'] - df_copy['prediction']
df_copy['error_abs'] = np.abs(df_copy['error'])

df_copy['error_perc'] = df_copy['error_abs']/df_copy['bean_high']
```

```python
df_copy[['Brazil_Production','bean_high','prediction', 'error', 'error_abs','error_perc']].head(50)
```

```python
#mean_absolut_error
mae = np.mean(df_copy['error_abs'])
```

```python
print('MAE = {}'.format(mae))
```

```python
#mean_absolut_Percentage_error
mape = np.mean(df_copy['error_perc'])
print("MAPE: {}".format( mape ) )
```
